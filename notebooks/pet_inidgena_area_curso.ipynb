# Importação de Bibliotecas

```python
import pandas as pd
```

# Importação da Base de Dados

```python
dados = pd.read_csv('../dados_pet_indigena.csv')
dados.info()
```

```python
dados.head()
```

```python
dados['ÁREA GERAL'] = dados['ÁREA GERAL'].str.strip()
```

# Visualizações

## Número discentes por área



```python
dados_area = dados.groupby(['CAMPUS', 'ÁREA GERAL'])['QUANTITATIVO DISCENTE INDÍGENA ATIVO GRADUAÇÃO'].sum().reset_index()
dados_area.to_csv('dados.csv', index=False)
dados_area
```

```python
dados_area = dados_area.pivot_table(index='ÁREA GERAL', columns='CAMPUS', values='QUANTITATIVO DISCENTE INDÍGENA ATIVO GRADUAÇÃO').reset_index()
# Soma das colunas especificadas
dados_area['TOTAL'] = dados_area[['EAD', 'I', 'II', 'III', 'IV']].sum(axis=1)

# Ordenação do DataFrame pela soma total
dados_area = dados_area.sort_values(by='TOTAL', ascending=False)
dados_area
```

```python
dados_area.to_csv('dados_area.csv', index=False)
```
